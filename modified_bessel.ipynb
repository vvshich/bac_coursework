```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import iv, kv
r = np.linspace(0, 5, 100)
y1 = iv(0, r)
y2 = kv(0, r)
```

```python
plt.plot(r, y1)
plt.grid(True)
plt.title("Modified Bessel function of the first kind")
plt.xlabel('X')
plt.ylabel('Y')
ax = plt.gca()
ax.set_xlabel("X", fontsize=15, labelpad=0)
ax.set_ylabel("Y", fontsize=15, labelpad=10, rotation=0)
ax.yaxis.set_label_coords(-0.05, 0.95)
ax.xaxis.set_label_coords(1.005, -0.05)
plt.show()
```

```python
plt.plot(r, y2)
plt.grid(True)
plt.title("Modified Bessel function of the second kind")
plt.xlabel('X')
plt.ylabel('Y')
ax = plt.gca()
ax.set_xlabel("X", fontsize=15, labelpad=0)
ax.set_ylabel("Y", fontsize=15, labelpad=10, rotation=0)
ax.yaxis.set_label_coords(-0.05, 0.95)
ax.xaxis.set_label_coords(1.005, -0.05)
plt.show()
```
